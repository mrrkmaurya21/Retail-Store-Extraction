```python
#Import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# List to store details of the Retail store
Store_Name, Address, Timing, Coordinates, Phone_Number = list(), list(), list(), list(), list()

# The outputfile after extraction
output_file = 'output'

# URL used to get the information
base_url = 'https://stores.vmartretail.com/?page='

# Module to get the details from url
def scrape_page(url):
    # Send a GET request to the URL and get the response
    response = requests.get(url)
    
    # Create a BeautifulSoup object from the response text using the html.parser
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extract data from the page using BeautifulSoup selectors or regular expressions
    for soup1 in soup.find_all('div', class_="col-lg-4 col-md-6 p-0 my-3"):
        # Get the address of the store and append it to the Address list
        Address1 = soup1.find('span', class_='store-address-info idx-info-card-str-add-info')
        Address.append(Address1.text)
        
        # Get the store timings and append it to the Timing list
        time = soup1.find('span', class_='store-time-info idx-info-card-str-time-info')
        Timing.append(time.text)
        
        # Get the coordinates of the store by extracting the href attribute value of the direction link
        # Split the value by '/' and get the 6th element and then split it by '=' and get the 4th element
        direction = soup1.find('a', class_='get_direction_click btn w-100 idx-info-card-str-dir-btn txt-color-3 bdr-color-6 bg-color-4 text-uppercase btn-hover-2')['href']
        destination = direction.split('/')[5]
        coordinates = destination.split('=')[3]
        Coordinates.append(coordinates)
        
        # Get the phone number of the store, split the string by space and append it to the Phone_Number list
        Mobile_no = soup1.find('a', class_="txt-hover-1 txt-color-1").text
        Mobile = Mobile_no.split(' ')
        Mobile_No = Mobile[0] + Mobile[1] + Mobile[2]
        Phone_Number.append(Mobile_No)
        
        # Get the name of the store and append it to the Store_Name list
        Store_Name.append(soup1.h2.a.text)
    
    # Create a Pandas DataFrame with the lists and write it to a CSV file
    df = pd.DataFrame({'Store Name': Store_Name, 'Address': Address, 'Timing': Timing, 'Coordinates': Coordinates, 'Phone Number': Phone_Number})
    df.to_csv(output_file+'.csv', index=False, encoding='utf-8')
    
    # Return the DataFrame
    return df

# Define the total number of pages to scrape
response = requests.get('https://stores.vmartretail.com/')
soup2 = BeautifulSoup(response.text, 'html.parser')
num_pages = max_pages = int((soup2.find('a', title='Last')['href'].split('='))[1])

# Scrape each page in the range of page numbers
for page_num in range(1, num_pages + 1):
    url = base_url + str(page_num)
    # Save the data to a file or database by calling the scrape_page function
    scrape_page(url)
```
